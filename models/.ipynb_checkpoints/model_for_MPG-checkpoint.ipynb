```python
import pandas as pd
import numpy as np
from sklearn import pipeline,preprocessing,metrics,model_selection,ensemble
from sklearn_pandas import DataFrameMapper
from sklearn.preprocessing import OneHotEncoder
```

```python
data=pd.read_excel(r'G:\DBA\MOD_2_core analytics\django\djangoTut\mpgWebApp\models\mileage_new.xls')
```

```python
data.head()
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
mapper = DataFrameMapper([
                        (['cyl','disp','hp','wt','acc','modyr'],preprocessing.StandardScaler()),
])
```

```python
pipeline_obj=pipeline.Pipeline([
    ("mapper",mapper),
    ("model",ensemble.RandomForestRegressor())
])
```

```python
data.columns
```

```python
X=['cyl', 'disp', 'hp', 'wt', 'acc', 'modyr']
Y=['mpg']
```

```python
pipeline_obj.fit(data[X],data[Y])
```

```python
pipeline_obj.predict(data[X])
```

```python
import joblib
```

```python
joblib.dump(pipeline_obj,'RFModelforMPG5.pkl')
```

```python
modelReload=joblib.load('RFModelforMPG5.pkl')
```

```python
modelReload.predict(data[X])
```
